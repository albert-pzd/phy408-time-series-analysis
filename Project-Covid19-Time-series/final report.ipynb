```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = pd.read_csv('countries.csv', parse_dates=['Date'])
data2 = pd.read_csv('key-countries-pivoted_csv.csv', parse_dates=['Date'])
```

```python
canada_data = data1[data1['Country'] == 'Canada'][0:-1]
china_us_data = data2[['Date', 'China', 'US']]
full_data = pd.merge(canada_data, china_us_data, on = 'Date')[['Date', 'China', 'US', 'Confirmed']]
full_data.columns = ['Date', 'China', 'US', 'Canada']
#full_data = full_data.set_index('Date')
full_data.set_index('Date',inplace=True, drop=True)
full_data.index = pd.to_datetime(full_data.index)
full_data.plot()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(full_data.set_index("Date"))
```
